```python
import pandas as pd
import matplotlib as plot
import matplotlib.pyplot as plt
df=pd.read_csv("rottenTomatoesUserReviews.csv",encoding="cp437")
```

```python
# Make the histogram using of No. of different Ratings given to different movies             
mnh=[]
names=[]
for movie in df['Movie'].unique():
    mnh.append(list(df[df.Movie == movie]['Rating']))
    names.append(str(movie))
colors = ['#006400','#FF69B4','#FFC0CB','#00BFFF','#4B0082','#000000','#7FFFD4',
          '#32CD32','#FF0000','#FF8C00','#0000FF','#FFFF00','#00FF00','#20B2AA','#FFD700','#FF4500']
plt.figure(figsize=(29,5))
                                                 
plt.hist(mnh, bins = int(180/90),color = colors, label=names)
plt.legend()
plt.xlabel('Rating')
plt.title('Side-by-Side Histogram with Multiple Movies')
```

```python
#Density curve for ratings distribution
df.Rating.plot.density()
```

```python
#pie Chart for different movies
mlist=list(df.Movie.unique())                          
series=pd.Series(df.Movie.value_counts(),index=mlist,name='series')
series.plot.pie(figsize=(9,9))
```
